### We are required to Proximity of Two Binary Object Vectors with Simple Matching, Jaccard Similarity, Cosine Similarity

```python
def simple_matching(p1,p2,cols):
    matched=0
    for i in cols:
        if p1.loc[i]==p2.loc[i]:
            matched+=1
    return matched

def jaccard_similarity(p1,p2,cols):
    total=len(cols)
    match=simple_matching(p1,p2,cols)
    return match/total

def cosine_similarity(p1,p2,cols):
    numerator= np.dot(p1.loc[cols], p2.loc[cols])
    mod_p1 = np.sqrt(sum(p1.loc[cols]))
    mod_p2 = np.sqrt(sum(p2.loc[cols]))
    return numerator/(mod_p1 * mod_p2)
    
    
```

### Correlation Analysis for Two Features
1.	Calculate Correlation between two features Age and Yearly Income 
2.	Calculate Correlation between two features Commute Distance and Yearly Income


```python
def correlation(f1,f2):
    m1=np.mean(f1)
    m2=np.mean(f2)
    std1=np.std(f1)
    std2=np.std(f2)
    t=(f1-m1)*(f2-m2)
    t1=np.sum(t)
    cov= t1/(df2.shape[0])
    return cov/(std1*std2)
        
```

```python
correlation(df2['Age'],df2['YearlyIncome'])
```

```python
correlation(df2['CommuteDistance'],df2['YearlyIncome'])
```
